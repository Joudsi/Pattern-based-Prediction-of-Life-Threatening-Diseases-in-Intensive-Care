# Clustering for BB patients

```python
data_bb = pd.read_csv('/Users/joudsi/Desktop/subdata_bb.csv')
```

```python
data_bb.head()
```

```python
subdata_bb = data_bb[['amdission_age','gender', 'tempc_mean', 'spo2_mean', 'resprate_mean', 'creatinine_mean', 'potassium_mean','bicarbonate_mean', 'hematocrit_mean',
       'glucose_mean', 'lactate_mean', 'albumin_mean', 'hemoglobin_mean']]
subdata_bb_norm = (subdata_bb - subdata_bb.min()) / (subdata_bb.max() - subdata_bb.min())
```

### Elbow method to determine the best K

```python
from sklearn.cluster import KMeans
from sklearn import datasets
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(subdata_bb_norm)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

### Start clustering based on the best K-value

```python
from sklearn.cluster import KMeans
from sklearn import datasets
model=KMeans(n_clusters=3)
model.fit(subdata_bb_norm)
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
md=pd.Series(model.labels_)
subdata_bb_norm['clust']=md
subdata_bb_norm.head()
```

```python
subdata_bb_norm['is_AKI']= data_bb['is_AKI']
```

```python
subdata_bb_norm.head(10)
```

```python
import seaborn as sns
% matplotlib inline

sns.countplot(x="is_AKI", hue="clust",data=subdata_bb_norm)
```

```python
subdata_bb_norm.groupby('clust').mean()
```

```python
subdata_bb_norm.to_csv("/Users/joudsi/Desktop/subdata_bb_norm_clustered.csv")
```
